```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Define the tweet URL
url = 'https://twitter.com/westerosies/status/1632423392279183360'

# Define empty lists to store comments and replies
comments_list = []
replies_list = []

# Define a function to extract replies
def extract_replies(comment):
    for tweet in sntwitter.TwitterSearchScraper('to:' + comment.url).get_items():
        if tweet.inReplyToTweetId == comment.id:
            replies_list.append(tweet)
            extract_replies(tweet)

            # Scrape comments
for tweet in sntwitter.TwitterSearchScraper('url:' + url).get_items():
    if tweet.inReplyToTweetId is None:
        comments_list.append(tweet)
        extract_replies(tweet)

# Convert comments and replies lists to Pandas DataFrames
comments_df = pd.DataFrame(comments_list)
replies_df = pd.DataFrame(replies_list)

# Save comments and replies to CSV files
comments_df.to_csv('comments.csv', index=False)
replies_df.to_csv('replies.csv', index=False)
```
